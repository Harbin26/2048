```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import time
import warnings
warnings.filterwarnings("ignore")
##############################################
# Load your dataset
data_frame = pd.read_csv('Real estate.csv')

# Display the first few rows of the dataset
data_frame.head()
```

```python
# to check the list of columns in the dataset
#list(df.columns)
data_frame.info()
```

```python
# checking if any column data is empty
data_frame.notnull()
```

Missing values in attributes would mean a data entry error. verified that there are no missing values.

### Shortlisting of data

```python
#selecting the attributes
short_list_features = ['X2 house age', 'X4 number of stores', 'X5 latitude', 'X6 longitude', 'Y price sq feet']
print(short_list_features)
```

### (a) Mean, Median, and Standard Deviation:

```python
# Calculating mean, median, and standard deviation
data_frame.describe()
```

### (b)Boxplots:

```python
# Checking for any outliers
# Loop through each feature attributes and plot its boxplot
for feature in short_list_features:
    # Draw a boxplot for the current feature
    plt.figure(figsize=(5, 5))
    sns.boxplot(data=data_frame[feature])
    plt.title('Boxplot of {}'.format(feature))
    plt.xlabel(feature)
    plt.ylabel('Values')
    plt.tight_layout()
    plt.show()
```

```python
#To visualize in one graph
import seaborn as sns
import matplotlib.pyplot as plt

# Create a single plot with multiple boxplots
plt.figure(figsize=(10, 6))
sns.boxplot(data=data_frame[short_list_features])
plt.title('Boxplots of Features')
plt.xlabel('Features')
plt.ylabel('Values')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.show()
```

From the above, it is seen that there are outliers for longitude and latitude this is  because there is a presence of 0 in the longitude and latitude column attribute. Having zeros in latitude and longitude attributes is a problem because, in real scenarios, every location is identified with latitude and longitude hence having 0 means one such house/region never exists which is practically false thus having zeros in these attributes will be considered outliers and needs preprocessing.This can be verified the same from the data_frame.describe(), with min attribute

```python
# Replace zero values with mean value to address the outliers
data_frame['X5 latitude'] = data_frame['X5 latitude'].replace(0.000000, data_frame['X5 latitude'].mean())
data_frame['X6 longitude'] = data_frame['X6 longitude'].replace(0.000000, data_frame['X6 longitude'].mean())

for feature in short_list_features:
    # Draw boxplot for the current feature
    plt.figure(figsize=(5, 5))
    sns.boxplot(data=data_frame[feature])
    plt.title('Boxplot of {}'.format(feature))
    plt.xlabel(feature)
    plt.ylabel('Values')
    plt.tight_layout()
    plt.show()
```

```python
#single view after removing the outlier
import seaborn as sns
import matplotlib.pyplot as plt

# Create a single plot with multiple boxplots
plt.figure(figsize=(10, 6))
sns.boxplot(data=data_frame[short_list_features])
plt.title('Boxplots of Features')
plt.xlabel('Features')
plt.ylabel('Values')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.show()
```

### **Regression**

```python
# Scikit-Learn ≥0.20 is required
import sklearn
from sklearn import linear_model
# Select a linear model
import numpy as np
from sklearn.linear_model import LinearRegression

X= train_df['X2 house age']
y = train_df['Y price sq feet']
X2= np.array(X).reshape(-1,1) # reshape(-1,1) makes the array horizontal, y is vertical

# Train the model (we use all data for training -->
model = sklearn.linear_model.LinearRegression()
model.fit(X2, y)
r_sq = model.score(X2, y)
print(r_sq)
plt.scatter(X2, y, color='black', label='observed')

plt.plot(X2, model.predict(X2), label='fit', color='Green', linewidth=2)

plt.xlabel('house age')
plt.ylabel('price sq feet')
plt.title('Regression')
plt.legend(loc='best')

plt.show()
```

```python
# Make a prediction
X3 = test_df['X2 house age']
X4 = test_df['Y price sq feet']
X3 = np.array(X3).reshape(-1,1)
X4 = np.array(X4).reshape(-1,1)
print(X4[7])
# what is the charges for element 44 of the test bmi list.
price_sq_feet = model.predict([X3[7]])
print('house age = {:.2f}   price sq feet = {:.2f}'.format(X3[7][0], price_sq_feet[0]))
```
